```python
import matplotlib.pyplot as plt

# Data
students = ['A', 'B', 'C', 'D', 'E']
marks = [85, 78, 92, 74, 88]

# Plot
plt.bar(students, marks, color='skyblue', edgecolor='black')

# Labels and Title
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Bar Graph: Students vs Marks")
plt.tight_layout()

# Display Graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
subjects = ['Math', 'Science', 'English', 'History']
scores = [50, 60, 70, 80]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Plot
plt.pie(scores, labels=subjects, colors=colors, autopct='%1.1f%%',explode=(0, 0.1, 0, 0),shadow=True, startangle=90)
plt.title("Pie Chart: Subject-wise Score Distribution")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular

# Display Graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create x values
x = range(-10, 11)    # from -10 to 10

# Compute y = x^2
y = [i**2 for i in x]

# Plot
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y = x")
plt.title("Plot of y = x")
plt.grid(True)
plt.show()
```

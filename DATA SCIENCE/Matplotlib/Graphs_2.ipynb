```python
# A) Write a python program to generate a simple bar graph using matplotlib. The graph should be properly labeled.
# CODE:-

import matplotlib.pyplot as plt

students = ['A', 'B', 'C', 'D', 'E']
marks = [85, 78, 92, 74, 88]

plt.bar(students, marks, color="purple", label="Marks Scored")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
```

```python
# B) Write a python program to generate Pie-chart using matplotlib. The graph should be properly labeled.
# CODE:-
# Data
subjects = ['Math', 'Science', 'English', 'History']
scores = [90, 85, 70, 60]
colors = ['purple', 'grey', 'lightskyblue', 'black']

# Plot
plt.pie(scores, labels = subjects, colors = colors)
plt.title("Pie Chart: Subject-wise Score Distribution")

# Display Graph
plt.show()
```

```python
# C) Write a Python program to plot the function y = x2 using the matplotlib libraries.
# Code:-
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(-10, 10, 100)   # 100 points between -10 and 10

# Compute y = x^2
y = x ** 2

# Plot the function
plt.plot(x, y, color='blue', linewidth=2, label='y = x')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph of y = x")

# Add a legend
plt.legend()

# Display grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```

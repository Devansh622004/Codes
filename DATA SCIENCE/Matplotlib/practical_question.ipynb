```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(-5, 5, (2, 500))
y = np.random.uniform(-5, 5, (2, 500))

z = np.sqrt(x*x + y*y)

# Scatter plot with colormap based on z
plt.scatter(x, y, c=z, cmap='viridis')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot with Colormap Based on z')
plt.show()
```

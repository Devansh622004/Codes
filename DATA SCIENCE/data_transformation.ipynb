```python
import pandas as pd
```

```python
 df = pd.read_csv("data_cleaning_sample.csv")
```

```python
df
```

```python
# For Sorting the Values -> Ascending Order
df.sort_values("Join Date")
```

```python
# For Sorting the Values -> Descending Order
df.sort_values("Join Date", ascending=False)
```

```python
# If we want to sort the Table Acc. to both Join Date, Age
df2 = df.sort_values(["Join Date", "Age"]).copy()
```

```python
df2
```

```python
# If we want the index to start from 0 & be sequential, you can reset it using reset_index()
df2.reset_index(drop=True)
```

```python
# Ranking
df2["Rank"] = df2["Join Date"].rank(ascending=False)
```

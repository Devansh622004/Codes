```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a directed graph
G = nx.DiGraph()

# Add edges (example graph)
edges = [
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("B", "E"),
    ("C", "F"),
    ("E", "G"),
]
G.add_edges_from(edges)  # For Adding Edges

# DFS function (iterative using stack)
def dfs(graph, start):
    visited = set() # So that no node should be repeated.
    stack = [start]
    traversal_path = []

    while stack:
        node = stack.pop()  # Pop last inserted node
        if node not in visited:
            visited.add(node)
            traversal_path.append(node)
            # Add neighbors in reverse order to visit left-to-right
            neighbors = list(graph.neighbors(node))
            stack.extend(reversed(neighbors)) # fensures left-to-right order
    return traversal_path

# Run DFS starting from 'A'
start_node = "A"
dfs_path = dfs(G, start_node)

print("DFS Traversal Path:", dfs_path)

# Draw the graph
pos = nx.spring_layout(G)  # It arranges nodes in a visually balanced way using a force-directed algorithm.
plt.figure(figsize=(6, 4))
nx.draw(
    G, pos, with_labels=True, node_color="skyblue",
    node_size=2000, font_size=12, font_weight="bold", arrows=True
)
plt.title("Directed Graph - DFS Example")
plt.show()
```
